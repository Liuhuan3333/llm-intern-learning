## 列表推导

```python
# 普通写法
squares = []
for i in range(10):
    squares.append(i ** 2)
print(squares)
```

```python
# 列表推导写法
squares = [i ** 2 for i in range(10)]
print(squares)
```

```python
# 带条件的列表推导，只保留偶数的平方
even_squares = [i ** 2 for i in range(10) if i % 2 == 0]
print(even_squares)
```

## 生成器
生成器的关键是懒计算，不会一次性把所有结果存在内存里，处理大数据时很重要。

```python
# 生成器函数：用yield代替return
def infinite_counter(start=0):
    n = start
    while True:
        yield n
        n += 1

gen = infinite_counter()
print(next(gen))  # 0
print(next(gen))  # 1
print(next(gen))  # 2

# 生成器表达式（和列表推导很像，但用圆括号）
gen_exp = (i ** 2 for i in range(1000000))  # 不会立即计算100万个数
print(next(gen_exp))  # 只计算第一个：0
print(next(gen_exp))  # 再计算下一个：1
```

## 装饰器
装饰器是LLM框架源码里非常常见的模式，理解它很重要。

```python
import time

# 自己写一个计时装饰器
def timer(func):
    def wrapper(*args, **kwargs):
        start = time.time()
        result = func(*args, **kwargs)
        end = time.time()
        display(f"{func.__name__} 运行耗时：{end - start:.4f} 秒")
        return result
    return wrapper

@timer
def slow_function(n):
    total = 0
    for i in range(n):
        total += i
    return total

result = slow_function(1000000)
display(f"结果：{result}")
```

## 类与继承

```python
# 基类
class Animal:
    def __init__(self, name, sound):
        self.name = name
        self.sound = sound
    
    def speak(self):
        return f"{self.name} 说：{self.sound}"
    
    def __repr__(self):
        return f"Animal(name={self.name})"

# 子类继承
class Dog(Animal):
    def __init__(self, name):
        super().__init__(name, "汪汪")  # 调用父类的__init__
    
    def fetch(self, item):
        return f"{self.name} 捡回了 {item}"

class Cat(Animal):
    def __init__(self, name):
        super().__init__(name, "喵喵")
    
    def speak(self):  # 重写父类方法
        return f"{self.name} 优雅地说：{self.sound}~"

dog = Dog("旺财")
cat = Cat("咪咪")

display(dog.speak())
display(dog.fetch("球"))
display(cat.speak())
display(dog)
```

练习1：用NumPy手写矩阵乘法

```python
import numpy as np

def matrix_multiply(A, B):
    """
    手动实现矩阵乘法
    A: shape (m, n)
    B: shape (n, p)
    返回: shape (m, p)
    """
    m, n = A.shape
    n2, p = B.shape
    assert n == n2, "A的列数必须等于B的行数"
    
    C = np.zeros((m, p))
    for i in range(m):
        for j in range(p):
            for k in range(n):
                C[i][j] += A[i][k] * B[k][j]
    return C

# 测试
A = np.random.randint(1, 5, size=(3, 4)).astype(float)
B = np.random.randint(1, 5, size=(4, 2)).astype(float)

C_manual = matrix_multiply(A, B)
C_numpy = A @ B

display("手动结果：")
display(C_manual)
display("\nNumPy结果：")
display(C_numpy)
display("\n两者是否一致：", np.allclose(C_manual, C_numpy))
```

练习2：用Pandas分析泰坦尼克数据集

```python
import pandas as pd
import matplotlib.pyplot as plt

# 直接从网络加载泰坦尼克数据集
url = "https://raw.githubusercontent.com/datasciencedojo/datasets/master/titanic.csv"
df = pd.read_csv(url)

display("数据形状:", df.shape)
display("\n前5行：")
display(df.head())
display("\n基本信息：")
display(df.info())
display("\n数值列统计：")
display(df.describe())
```

```python
# 查看缺失值
display("各列缺失值数量：")
display(df.isnull().sum())
```

```python
# 分析1：总体生存率
survival_rate = df["Survived"].mean()
display(f"总体生存率：{survival_rate:.2%}")

# 分析2：不同性别的生存率
gender_survival = df.groupby("Sex")["Survived"].mean()
display("\n各性别生存率：")
display(gender_survival)

# 分析3：不同舱位等级的生存率
class_survival = df.groupby("Pclass")["Survived"].mean()
display("\n各舱位等级生存率：")
display(class_survival)

# 分析4：不同性别+舱位的组合分析
combined = df.groupby(["Sex", "Pclass"])["Survived"].mean().unstack()
display("\n性别×舱位 生存率矩阵：")
display(combined)
```

```python
# 可视化（保存为图片）
fig, axes = plt.subplots(1, 3, figsize=(15, 4))

# 图1：生存人数分布
df["Survived"].value_counts().plot(kind="bar", ax=axes[0], color=["salmon", "steelblue"])
axes[0].set_title("生存/死亡人数")
axes[0].set_xticklabels(["死亡", "生存"], rotation=0)

# 图2：各性别生存率
gender_survival.plot(kind="bar", ax=axes[1], color="steelblue")
axes[1].set_title("各性别生存率")
axes[1].set_xticklabels(["女性", "男性"], rotation=0)
axes[1].set_ylim(0, 1)

# 图3：年龄分布
df["Age"].dropna().hist(bins=30, ax=axes[2], color="steelblue", edgecolor="white")
axes[2].set_title("年龄分布")
axes[2].set_xlabel("年龄")

plt.tight_layout()
plt.savefig("titanic_analysis.png", dpi=120)
plt.show()
display("图表已保存")
```
